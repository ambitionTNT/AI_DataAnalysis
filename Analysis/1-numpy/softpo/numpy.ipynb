### 2. 部分排序
np.partition(a,k)

有的时候我们不是对全部数据感兴趣，我们可能只对最小或最大的一部分感兴趣。
- 当k为正时，我们想要得到最小的k个数
- 当k为负时，我们想要得到最大的k个数

```python
nd = np.random.randint(0, 1000, size = 100)
```

```python
np.partition(nd, kth = 10)
```
