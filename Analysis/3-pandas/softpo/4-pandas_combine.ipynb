## 2. 使用pd.merge()合并

merge与concat的区别在于，merge需要依据某一共同的行或列来进行合并

使用pd.merge()合并时，会自动根据两者相同column名称的那一列，作为key来进行合并。

注意每一列元素的顺序不要求一致

###  1) 一对一合并

````python
df1 = DataFrame({'id': [1, 10, 1024], 'name': ['po','alisa','Jack'], 'sex': ['male', 'female', 'male']})
df1 
````

````python
df2 = DataFrame({'id':[1, 11, 111], 'salary': [10000, 20000, 30000], 'age':[20, 30, 40]})
df2
````

````python
df1.merge(df2)
````

### 2) 多对一合并

````python
df3 = DataFrame({'id':[1, 1, 1], 'salary': [10000, 20000, 30000], 'age':[20, 30, 40]})
df3
````

````python
df1.merge(df3)
````

### 3) 多对多合并

````python
df3 = DataFrame({'id':[1,1,1], 'name': ['po','alisa','Jack'], 'sex': ['male', 'female', 'male']})
df3
````

````python
pd.merge(df2, df3)#经典笛卡尔积 
````

### 4) key的规范化

- 使用on=显式指定哪一列为key,当有多个key相同时使用

````python
df1 = DataFrame({'id': [1, 10, 1024], 'name': ['po','alisa','Jack'], 'sex': ['male', 'female', 'male']})
df1 
df2 = DataFrame({'id':[1, 11, 111], 'name': ['hello', 'world', 'sb'], 'age':[20, 30, 40]})
df2
display(df1,df2)
````

````python
pd.merge(df1,df2,on='id')
````

- 使用left_on和right_on指定左右两边的列作为key，当左右两边的key都不想等时使用

````python
df2 = DataFrame({'ID':[1, 11, 111], 'salary': [10000,20000,30000], 'age':[20, 30, 40]})
df2
````

````python
pd.merge(df1,df2, left_on = 'id', right_on = 'ID')
````

### 5) 内合并与外合并

- 内合并：只保留两者都有的key（默认模式）

````python
display(df1, df2)
````

````python
pd.merge(df1, df2, how = 'inner',left_on = 'id', right_on = 'ID')#默认内合并
````

- 外合并 how='outer'：补NaN

````python
'''how:{left,right,outer,inner}'''
pd.merge(df1, df2, how = 'outer',left_on = 'id', right_on = 'ID')#外合并
````

- 左合并、右合并：how='left'，how='right'，

````python
pd.merge(df1, df2, how = 'left',left_on = 'id', right_on = 'ID')#左合并，左边所有的数据为主表
````

````python
pd.merge(df1, df2, how = 'right',left_on = 'id', right_on = 'ID')#左合并，左边所有的数据为主表
````

### 6) 列冲突的解决

当列冲突时，即有多个列名称相同时，需要使用on=来指定哪一个列作为key，配合suffixes指定冲突列名

可以使用suffixes=自己指定后缀

````python
df1 = DataFrame({'id': [1, 10, 1024], 'name': ['po','alisa','Jack'], 'sex': ['male', 'female', 'male']})
df1 
df2 = DataFrame({'id':[1, 11, 111], 'name': ['hello', 'world', 'sb'], 'age':[20, 30, 40]})
df2
display(df1,df2)
````

````python
pd.merge(df1, df2, on = 'id', suffixes = ('_A','_B'))
````
