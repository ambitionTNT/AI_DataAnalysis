## 5. 数据聚合【重点】

数据聚合是数据处理的最后一步，通常是要使每一个数组生成一个单一的数值。

数据分类处理：

 - 分组：先把数据分为几组
 - 用函数处理：为不同组的数据应用不同的函数以转换数据
 - 合并：把不同组得到的结果合并起来
 
数据分类处理的核心：
     groupby()函数

```python
df = DataFrame({'item': np.random.randint(0,10,size = 100),
               'saler': np.random.randint(0, 10, size = 100),
               'weight':np.random.randint(30, 300, size  = 100),
                'price':np.random.randint(1, 20, size = 100)
               })
df
```

```python
def convert_item(x):
    if x< 2:
        return '萝卜'
    elif x<7:
        return '白菜'
    else:
        return '青椒'
df['item'] = df['item'].map(convert_item)
```

```python
def convert_saler(x):
    if x<5:
        return '张传龙'
    elif x<7:
        return '孙悟空'
    else:
        return '唐僧'
df['saler'] = df['saler'].map(convert_saler)
```

```python
def convert_weight(x):
    if x< 80:
        return 50
    elif x <150:
        return 130
    else :
        return 150
df['weight'] = df['weight'].transform(convert_weight)
```

```python
df
```

```python
#统计一月的销售重量
all_sale = df.groupby(['saler'])['weight']
all_sale
```

```python
#分组之后需要计算
all_sale.sum()
```

```python
all_sale.min()
```

```python
all_sale.apply(np.sum)
```

```python
all_sale.count()
```

```python
all_sale
```

```python
def count(x):
    return x.size
```

```python
all_sale.apply(count)
```

```python
def analysis(x):
    return (x.size, x.sum(), x.mean(), x.min)
all_sale.apply(analysis)
```
